# **Análisis de Vehículos Eléctricos en España**
---

### ***Introducción***


Este proyecto se centra en el **análisis de la evolución y situación actual de los vehículos eléctricos matriculados en España**, cubriendo aspectos clave como la *disponibilidad de puntos de carga, la comparativa con otros países europeos y el análisis de factores económicos como el precio y la autonomía de estos vehículos* y así también *poner en situación a los futuros y posibles compradores de estos vehiculos, las ventajas y desventajas en el panorama actual y conocer la rentabilidad de adquirir un eléctrico.*


El proyecto se enmarca en los esfuerzos de sostenibilidad de la **Agenda 2030**, que busca reducir la huella de carbono y promover un entorno urbano limpio y accesible. 

Se tienen en cuenta algunos **Objetivos de Desarrollo Sostenible (ODS)**, específicamente los ODS 7, 11 y 13, lo cual nos hace preguntarnos lo primero:

-   **¿Qué es la Agenda 2030 y sus ODS?**

#### **1.	Contexto de los ODS y su Relación con el Proyecto**

**La Agenda 2030 para el Desarrollo Sostenible**, también conocida como los **Objetivos de Desarrollo Sostenible (ODS)**, es un plan de acción adoptado por las Naciones Unidas en **2015**, **adquirido por 193 países miembros de la ONU**, que establece **17 objetivos y 169 metas** para abordar los desafíos globales más importantes *hasta el año 2030*.

Este compromiso universal **busca lograr un desarrollo sostenible económico, social y ambiental** y algunos de los objetivos abarcan áreas como:

- Erradicar la pobreza y la desigualdad
- Garantizar la educación y la salud para todos
- Proteger el planeta y combatir el cambio climático
- Fomentar la igualdad de género y el empoderamiento de las mujeres y las niñas
- Promover la paz y la reducción de la violencia
- Mejorar la calidad de vida y la sostenibilidad en las ciudades


La Agenda 2030 **es un llamamiento a la acción para todos los países, organizaciones y actores, para trabajar juntos en la implementación de estas metas y alcanzar un futuro más sostenible y próspero para todas las personas.**
________________________________________
- ***Los objetivos específicos que se han considerado para este proyecto son:***

**ODS 7 (Energía asequible y no contaminante)**: La transición hacia vehículos eléctricos requiere un suministro energético sostenible, por lo que el despliegue de puntos de carga y el acceso a energías renovables juegan un rol clave.


**ODS 11 (Ciudades y comunidades sostenibles)**: La movilidad eléctrica reduce la contaminación en zonas urbanas, contribuyendo a un entorno más saludable y menos dependiente de combustibles fósiles.


**ODS 13 (Acción por el clima)**: El cambio reducirá las emisiones de gases de efecto invernadero, alineándose con las metas globales de mitigación del cambio climático.


Estos objetivos son clave para enmarcar el análisis y evaluar el impacto de la movilidad eléctrica en España, proporcionando una perspectiva de sostenibilidad y un contexto relevante para interpretar los resultados. Tambien **El Plan Nacional Integrado de Energía y Clima (PNIEC)** 2021-2030 de España ha realizado actualizaciones significativas fortaleciendo objetivos e **incluyendo la meta de 5,5 millones de Vehículos Eléctricos para 2030.**

#### **2.	Análisis de la Evolución de los Vehículos Eléctricos en España (2020-2024)**


Para comprender el crecimiento de los vehículos eléctricos en España, se han recogido datos del *portal de la **DGT***, se ha realizado un trabajo para la transformación de estos y así tratarlos posteriormente.


Una vez listos, se ha filtrado directamente por los que tienen el código **“2”** pertenecientes al tipo de **“Eléctricos”**. Se ha explicado qué tipos diferentes de vehículos existen en el rango de fechas y su evolución.
*Llegados a este punto **se plantea la pregunta de:***

- **¿Cómo se ve España ante el tema de matriculaciones de V.E.?**
- **¿Y su evolución en los últimos cuatro años?**
 
Para profundizar en las razones detrás de esta evolución, **se consideraron factores como los incentivos gubernamentales, las subvenciones europeas (Plan MOVES III), y el impulso hacia los ODS.**


A lo largo de los años, la implementación de planes como el **Plan MOVES en España** ha proporcionado ***subvenciones para la adquisición de vehículos eléctricos***, lo que ha contribuido al aumento en el número de matriculaciones. Además, las políticas europeas de reducción de emisiones y el compromiso con la electrificación del transporte jugaron un papel importante en esta tendencia.

#### **3.   Comparativa de Vehículos Eléctricos y Análisis de Puntos de Carga entre Países**

- **¿Y respecto a Europa, en que punto nos encontramos tanto en matriculaciones como en los puntos de carga?**

Para ofrecer un contexto más amplio, se compararon los datos de **vehículos unicamente eléctricos BEV** y **puntos de carga** de *España* con otros países europeos, incluyendo Noruega, Francia, Alemania y Países Bajos. Este análisis resaltó cómo algunos países, como Alemania y Paises Bajos, han avanzado significativamente en la transición hacia la movilidad eléctrica, en parte debido a políticas más agresivas y una infraestructura de carga más desarrollada.

***Uno de los componentes más importantes*** para la adopción de vehículos eléctricos es la **infraestructura de carga**. Entonces: 
-   **¿tiene España una Infraestructura sólida de Puntos de Carga?**

Se utilizaron gráficos de barras y de mapas interactivos para ilustrar estas comparativas (y responder preguntas) de forma visualmente atractiva y comprensible, resaltando las diferencias y similitudes primeramente **a nivel nacional** y posterior **a nivel Europeo.**

#### **4. Panorama actual de Precios de Vehículos Eléctricos en España**

- **¿Existe un Programa de Ayudas impulsado por el Gobierno?**

Otro aspecto fundamental en el proyecto fue la **actualidad de los precios de los vehículos BEV**. Se analizaron diferentes modelos de **vehículos BEV** en el mercado español, comparando precio original con los precios ajustados bajo el **Plan MOVES III**, tanto con como sin ayuda por achatarramiento, que básicamente es si el comprador entrega o no un vehículo de más de siete años para ser desguazado, si lo entrega recibe una ayuda superior todavía. 

Se utilizaron gráficos de barras para ilustrar esta comparación, que permite observar de manera clara las reducciones en los precios gracias a los incentivos y ayuda a entender la rentabilidad y la viabilidad económica de los vehículos eléctricos resaltando el impacto positivo de las ayudas.

________________________________________
________________________________________
#### **5. Análisis de Autonomía de los Vehículos Eléctricos**

Otro aspecto relevante para los consumidores es la autonomía de los vehículos eléctricos por lo que se analizan **distintos modelos y sus autonomías**, tanto en el **ciclo WLTP** *(autonomía oficial, pero idealista)* como en la **autonomía estimada** en condiciones reales *(que es la que los usuarios suelen observar).*

Los gráficos de líneas muestran las variaciones de autonomía entre diferentes marcas y modelos, permitiendo una comparativa rápida y efectiva. Este análisis es crucial para evaluar la usabilidad y practicidad de los vehículos eléctricos en España, considerando que las distancias y el acceso a puntos de carga varían según la región.

#### **6. Rentabilidad de los Vehículos Eléctricos**

- **¿Cuánta rentabilidad tiene poseer un vehículo eléctrico hoy en día?**

Por último, se extraen conclusiones sobre la rentabilidad de los vehículos eléctricos frente a los vehículos de combustión:

- Incentivos económicos para la compra de vehículos eléctricos, 
- Aumento de la disponibilidad de puntos de carga en áreas urbanas y rurales, 
- El compromiso con los objetivos climáticos 
- El fomento de un estilo de vida sostenible y ciudades menos contaminadas.

#### Dificultades Encontradas

Durante el proyecto tuve dificultades como:

1.	Gestión y limpieza de datos: Algunos datasets contenían valores inconsistentes o formatos complejos, como el dataset de la DGT, al cual tuve que dedicar el mayor tiempo a transformar a un formato legible, limpiar y volver a transformar a CSV para posteiormente comenzar el análisis. Otros fueron pobres de datos y opté por recopilar informacion de forma más manual y con la ayuda de la IA, completando asi un conjunto de CSV para mi análisis.

2.	Representar gráficos: Algunos de ellos, como los precios de los vehículos agrupados por marca, y tambien el conseguir un estilo personalizado para todos ellos fue lo más desafiante. Pero el fallo error hace que el uso de visualizar gráficos con Plotly y Folium sea más familiar y ameno para la proxima exploración de datos

3.	Alcance y relación con los ODS: Encontrar una manera de alinear el proyecto con los objetivos de la Agenda 2030 no fue inmediato. Requirió un esfuerzo adicional, practicamente hasta los últimos dias, para investigar los ODS relacionados y adaptar el enfoque del análisis a estos objetivos.
